```python
def rpca_gd(M, lambda_1, lambda_2, maxiter, alpha):
    # Initialize U, V, S
    m, n = M.shape[0], M.shape[1]
    k = min(m, n)
    U = np.random.normal(0, 1, (k, m))
    V = np.random.normal(0, 1, (k, n))
    
    S = np.sign(np.random.normal(0, 1, (m, n)))
    
    
    
    # Gradient descent update
    iter_ = 0
    
    while iter_ < maxiter:  
        err = -2* (M - (U.T@V))
        print(np.linalg.norm(err))
#         grad_S = err + np.sign(S)*lambda_2
        grad_U = V@err.T + lambda_1*U
        grad_V = U@err + lambda_1*V
        
        U = U - alpha*grad_U
        V = V - alpha*grad_V
#         S = S - alpha*grad_S
        
        iter_ += 1
        
    L = U.T@V
    
    return L, S
        
        
        

                
    
```
